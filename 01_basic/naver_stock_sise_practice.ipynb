```python
import time
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import openpyxl
import requests
```

```python
url='https://finance.naver.com/item/sise.naver?code=005930'
html=requests.get(url).text

bs=BeautifulSoup(html,"html5lib")
```

```python
# 태그 이름으로 특정하거나 태그의 class로 특정하거나(이 때는 .붙임)
table=bs.select_one('.type_tax')
print(table)
```

```python
# 위 자료에서 class 이름이 "tah p11" 로 되어있음.
# 이렇게 띄어쓰기로 결합된 경우에는 클래스명을 2개라고 보면 됨.
# 즉 tah만 쓰거나 p11만 쓰거나
tags=table.select('.p11')
tags
```

```python
# select_one()은 첫번째 결과인 반면 select()는 전수 반환
nametags=table.select('.title')
```

```python
total=[]

for t,n in zip(tags,nametags):
    temp=t.text
    name=n.text
    temp=temp.replace('\n','')
    temp=temp.replace('\t','')
    total.append({name:temp})

total
```
