```python
import crawler
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from selenium import webdriver
from bs4 import BeautifulSoup
import openpyxl
```

```python
# 엑셀 시트 만들어놓기
wb=openpyxl.Workbook()
sheet=wb.active
sheet.append(["종목명",'현재가'])
```

```python
wd=crawler.call_chrome()
url='https://finance.naver.com/sise/sise_quant.naver'
wd.get(url)
time.sleep(2)

bs=BeautifulSoup(wd.page_source,"lxml")
top_trading=[]

# find_all은 조건에 맞는 모든 tag 리턴
# attrs는 태그 내에 추가조건 설정할 때 사용하고 딕셔너리 형태로 입력
table_box=bs.find("table", attrs={"class":"type_2"})
stock_name_list=table_box.findAll("a", attrs={"class":"tltle"})
stock_price_list=table_box.findAll("td", attrs={"class":"number"})

for i in range(100):
    stock_name=stock_name_list[i].get_text()
    stock_price=stock_price_list[i*10].get_text()
    sheet.append([stock_name, stock_price])
    top_trading.append({"종목명":stock_name,
                       "현재가":stock_price})

print(top_trading)
wd.quit()
```

```python
wb.save("네이버 거래상위 주식.xlsx")
stock_file=pd.DataFrame(top_trading)
stock_file.to_csv("네이버 거래상위 주식.csv", encoding='euc-kr')
```
